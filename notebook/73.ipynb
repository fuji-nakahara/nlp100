```ruby
require 'daru'
require 'fast-stemmer'
require 'statsample-glm'

require_relative '../src/71'
```

```ruby
features = File.readlines('../output/72.txt', chomp: true)
features.first(10)
```

```ruby
cols = [:y] + features
rows = []
File.open('../output/sentiment.txt') do |file|
  file.each do |line|
    arr = Array.new(cols.size, 0)
    y, *words = line.scrub.split
    arr[0] = 1 if y == '+1'
    words.reject(&method(:stopword?)).map(&:stem).uniq.each do |word|
      index = cols.index(word)
      arr[index] = 1 if index
    end
    rows << arr
  end
end
dataset = Daru::DataFrame.rows(rows, order: cols)
dataset.head
```

```ruby
glm = Statsample::GLM.compute(dataset, :y, :logistic, constant: 1)
coefficients = glm.coefficients(:hash)
```
